1. 日本語のテキストマイニングについて調べて下さい。

解答：日本語のテキストマイニングは特に、言語の複雑さ（たとえば、漢字、ひらがな、カタカナの混在、膨大な数の表現形式、文脈による意味の変化など）を扱うことに特化している。

2. 形態素解析について調べて下さい。

解答：形態素解析とは、文を言葉の最小単位である形態素に分割し、それぞれの品詞情報を付与するプロセスである。これにより、テキストの構造を理解しやすくなり、後続の分析作業が容易になる。
日本語には単語の間にスペースがなく、文を形態素に分割することが自然言語処理の大切なフェーズである。

6. Sudachi を python で実装して下さい。

---

```python
from sudachipy import tokenizer
from sudachipy import dictionary

# Sudachiのトークナイザーを設定
tokenizer_obj = dictionary.Dictionary().create()

# 分析したいテキスト
text = "Sudachiは日本語の形態素解析のためのツールです"

# 形態素解析の実行
# SplitModeを指定する場合は tokenize メソッドのオプションとして渡します
tokens = tokenizer_obj.tokenize(text, tokenizer.Tokenizer.SplitMode.C)

# 結果の表示
for token in tokens:
    print(token.surface(), token.part_of_speech())

```

7. Janome を使う場合に、ストップワードを取り除くコードを実装して下さい。

```python
from janome.tokenizer import Tokenizer

tokenizer = Tokenizer()
text = "JanomeはPythonで書かれた日本語の形態素解析器です。"

stopwords = ["は", "で", "た", "の", "です", "。"]

tokens = tokenizer.tokenize(text)

filtered_tokens = []

for token in tokens:
    if token.surface not in stopwords:
        filtered_tokens.append(token)
        
for token in filtered_tokens:
    print(token.surface) 
```

10. headline と bodysub のそれぞれに対して、名詞を抽出してワードクラウドを作って下さい。
---

```python
from janome.tokenizer import Tokenizer
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Janomeのトークナイザーのインスタンス生成
tokenizer = Tokenizer()

# 名詞を抽出する関数
def extract_nouns(text):
    return [token.surface for token in tokenizer.tokenize(text) if token.part_of_speech.startswith('名詞')]

# headlineとbodysubから名詞を抽出
nouns_headline = sum(df_e['headline'].apply(extract_nouns), [])
nouns_bodysub = sum(df_e['bodysub'].apply(extract_nouns), [])

# 名詞の出現回数をカウント
nouns_count_headline = Counter(nouns_headline)
nouns_count_bodysub = Counter(nouns_bodysub)

# ワードクラウドの設定
wordcloud = WordCloud(
    font_path='/Library/Fonts/Arial Unicode.ttf', 
    background_color='white', 
    width=800, 
    height=400
    )

# headlineのワードクラウド生成
wordcloud.generate_from_frequencies(nouns_count_headline)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# bodysubのワードクラウド生成
wordcloud.generate_from_frequencies(nouns_count_bodysub)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

11. headline と bodysub のそれぞれに対して、専門用語やキーフレーズ（複合語または名詞句）を抽出してワードクラウドを作って下さい。

---

```python
from janome.tokenizer import Tokenizer
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Janomeのトークナイザーをインスタンス化
tokenizer = Tokenizer()

"""
1.文章を単語に分けます。
2.それぞれの単語を見て、名詞かどうかをチェックします。
3.名詞ならば、それを「フレーズ」という箱に入れます。
4.名詞でない単語が来たら、その箱に入っている名詞たちをリストに加えます（これが一つの大切なフレーズです）。
5.そしてまた新しい名詞を探し始め、新しい「フレーズ」を作ります。
6.文章の最後までこの作業を続けます。
"""

# キーフレーズを抽出する関数
def extract_keyphrases(text):
    # 形態素解析を実行 (名詞のみ抽出)
    tokens = tokenizer.tokenize(text)
    # 名詞の連続する単語を抽出 (キーフレーズとして抽出) 
    keyphrases = []
    # "" は空文字列 <- フレーズを一時的に保存する変数,なぜ
    current_phrase = ""
    
    # 形態素解析の結果を1つずつ見ていく 
    for token in tokens:
        # 名詞が連続する場合はフレーズに追加
        if token.part_of_speech.startswith('名詞'):
            # フレーズに追加 += は文字列の連結, token.surface は形態素解析の結果の表層形
            current_phrase += token.surface
        else:
            if current_phrase:
                # フレーズがあればキーフレーズとして追加
                keyphrases.append(current_phrase)
                # フレーズを一時的に保存する変数を初期化 (空文字列にする)
                current_phrase = ""
    
    # 最後のフレーズを追加
    if current_phrase:
        # フレーズがあればキーフレーズとして追加 (最後のフレーズ)
        keyphrases.append(current_phrase)
    
    return keyphrases

# headlineとbodysubからキーフレーズを抽出, sum はリストの連結, [] は空のリスト 
keyphrases_headline = sum(df_e['headline'].apply(extract_keyphrases), [])
keyphrases_bodysub = sum(df_e['bodysub'].apply(extract_keyphrases), [])

# キーフレーズの出現回数をカウント
keyphrases_count_headline = Counter(keyphrases_headline)
keyphrases_count_bodysub = Counter(keyphrases_bodysub)

# ワードクラウドの設定
wordcloud = WordCloud(
    font_path='/Library/Fonts/Arial Unicode.ttf', 
    background_color='white', 
    width=800, 
    height=400
    )

# headlineのワードクラウド生成
# generate_from_frequencies は辞書型のデータを受け取ってワードクラウドを生成する
wordcloud.generate_from_frequencies(keyphrases_count_headline)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# bodysubのワードクラウド生成
# generate_from_frequencies は辞書型のデータを受け取ってワードクラウドを生成する
wordcloud.generate_from_frequencies(keyphrases_count_bodysub)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

すべての名詞（単一の名詞も含む）をキーワードとして抽出してしまった。
その結果、単一の名詞もフレーズとしてカウントされ、ワードクラウドに表示されてしまった

連続する名詞のみを含む名詞句をキーワードとして抽出し、単一の名詞を除外する処理を加える

```python

def extract_improved_keyphrases(text):
    
    tokens = tokenizer.tokenize(text)
    keyphrases = []
    current_phrase = ""
    current_length = 0  # 現在のフレーズの長さを追跡するための変数

    for token in tokens:
        # 名詞が連続する場合はフレーズに追加
        if token.part_of_speech.startswith('名詞'):
            current_phrase += token.surface
            current_length += 1
        else:
            # フレーズの最小長を2とする
            if current_length > 1:
                keyphrases.append(current_phrase)
            # フレーズをリセット
            current_phrase = ""
            current_length = 0
    
    # 最後のフレーズを追加
    if current_length > 1:
        keyphrases.append(current_phrase)
    
    return keyphrases

# 改善された関数を使ってキーフレーズを抽出
improved_keyphrases_headline = sum(df_e['headline'].apply(extract_improved_keyphrases), [])
improved_keyphrases_bodysub = sum(df_e['bodysub'].apply(extract_improved_keyphrases), [])

# キーフレーズの出現回数をカウント
improved_keyphrases_count_headline = Counter(improved_keyphrases_headline)
improved_keyphrases_count_bodysub = Counter(improved_keyphrases_bodysub)

# headlineのワードクラウド生成
wordcloud.generate_from_frequencies(improved_keyphrases_count_headline)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# bodysubのワードクラウド生成
wordcloud.generate_from_frequencies(improved_keyphrases_count_bodysub)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
